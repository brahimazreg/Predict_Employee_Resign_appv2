```python
preprocessor = ColumnTransformer(
    transformers=[
        (
            'cat',
            OneHotEncoder(handle_unknown='ignore'),
            categorical_features
        )
    ],
    remainder='passthrough'
)
```

```python
pipeline = Pipeline(
    steps=[
        ('preprocess', preprocessor),
        ('smoteenn', SMOTEENN(random_state=42)),
        ('model', DecisionTreeClassifier(
            criterion='gini',
            max_depth=6,
            min_samples_leaf=8,
            random_state=42
        ))
    ]
)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
pipeline.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = pipeline.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
y_pred[:10]
```

#### Class 0 (No Attrition)
#### ✔ Model is very good at predicting employees who stay
#### ✔ Most “No” predictions are correct
#### Class 1 (Attrition - IMPORTANT CLASS)
❗ Recall = 0.41
Model only catches 41% of employees who actually leave
 It misses 59% of real attrition cases
 Precision = 0.27
 When model predicts “Yes”, it's correct only 27% of the time
Many false alarms

```python
# I will tray an other model : Random Forest
from sklearn.ensemble import RandomForestClassifier

model_RF = RandomForestClassifier(
    n_estimators=200,
    max_depth=8,
    class_weight='balanced',
    random_state=42
)
```

```python
pipeline_rf=Pipeline(
    steps = [
            ('preprocess',preprocessor),
            ('smoteenn', SMOTEENN(random_state=42)),
            ('model_rf', RandomForestClassifier(
                         n_estimators=200,
                         max_depth=8,
                         class_weight='balanced',
                         random_state=42)
            )
])
```

```python
pipeline_rf.fit(x_train,y_train)
```

```python
y_pred_rf = pipeline_rf.predict(x_test)
print(classification_report(y_test, y_pred_rf))
```

### Conclusion :
#####  “The Random Forest model improved precision for attrition prediction from 27% to 42%, reducing false positives. However, recall decreased from
#####  42% to 33%, meaning more attrition cases were missed. This highlights the precision-recall tradeoff commonly encountered in balanced 
#####    classification ### problems.”
